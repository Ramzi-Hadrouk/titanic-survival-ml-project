```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data
df = pd.read_csv("../data/processed/train_processed.csv")
```

```python
# 1. Survival vs. Sex
survived_by_sex = df.groupby('Sex')['Survived'].mean()
plt.bar(survived_by_sex.index, survived_by_sex.values, color=['blue', 'pink'])
plt.title("Survival Rate by Sex")
plt.ylabel("Survival Rate")
plt.xlabel("Sex")
plt.show()
```

```python

# 2. Survival vs. Pclass
survived_by_pclass = df.groupby('Pclass')['Survived'].mean()
plt.bar(survived_by_pclass.index.astype(str), survived_by_pclass.values, color='orange')
plt.title("Survival Rate by Passenger Class")
plt.ylabel("Survival Rate")
plt.xlabel("Passenger Class")
plt.show()
```

```python

# 4. Survival vs. Family Size
survived_by_family = df.groupby('FamilySize')['Survived'].mean()
plt.bar(survived_by_family.index, survived_by_family.values, color='orange')
plt.title("Survival Rate by Family Size")
plt.ylabel("Survival Rate")
plt.xlabel("Family Size")
plt.xticks(survived_by_family.index)
plt.show()
```

```python
# 4. Survival vs. IsAlone
survived_by_IsAlone = df.groupby('IsAlone')['Survived'].mean()
plt.bar(survived_by_IsAlone.index.map({0: "Not Alone", 1: "Alone"}) ,survived_by_IsAlone.values, color=['black', 'gray'])
plt.title("Survival Rate by IsAlone")
plt.ylabel("Survival Rate")
plt.xlabel("IsAlone")
plt.xticks(survived_by_IsAlone.index)
plt.show()
```
